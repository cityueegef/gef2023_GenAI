```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Word vectors (3D made-up values such that cat and dog are closer)
word_vectors = {
    "Cat": np.array([1.0, 2.0, 1.5]),
    "Dog": np.array([1.5, 2.5, 1.8]),
    "Car": np.array([5.0, 5.0, 5.5])
}
#word_vectors = {
#    "Cat": np.array([1.0, 0.0, 0.0]),
#    "Dog": np.array([0.0, 1.0, 0.0]),
#    "Car": np.array([0.0, 0.0, 1.0])
#}

# Calculate Euclidean distances between the vectors
dist_cat_dog = np.linalg.norm(word_vectors["Cat"] - word_vectors["Dog"])
dist_cat_car = np.linalg.norm(word_vectors["Cat"] - word_vectors["Car"])
dist_dog_car = np.linalg.norm(word_vectors["Dog"] - word_vectors["Car"])

# Print distances to confirm
print(f"Distance between Cat and Dog: {dist_cat_dog:.2f}")
print(f"Distance between Cat and Car: {dist_cat_car:.2f}")
print(f"Distance between Dog and Car: {dist_dog_car:.2f}")

# Visualization in 3D
def plot_vectors_3d(vectors):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    for word, vec in vectors.items():
        ax.scatter(*vec, label=word)
        ax.text(vec[0] + 0.1, vec[1] + 0.1, vec[2] + 0.1, word, fontsize=12)

    # Set labels and title
    ax.set_title('Word Vector Visualization in 3D')
    ax.set_xlabel('x', fontsize=12, labelpad=15)
    ax.set_ylabel('y', fontsize=12, labelpad=15)
    ax.set_zlabel('z', fontsize=12, labelpad=15)
    #ax.view_init(elev=40, azim=300)
    plt.legend()
    plt.show()

# Plot the word vectors in 3D
plot_vectors_3d(word_vectors)
```
